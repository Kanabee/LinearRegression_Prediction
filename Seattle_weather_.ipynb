````python
import pandas as pd
import numpy as np
````

````python
data = pd.read_csv("seattle_weather_1948-2017.csv")
data
````

**Linear Regression Model**

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
````

Dummy [RAIN]

````python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data["RAIN_SUB"] = le.fit_transform(data["RAIN"])
````

````python
data
````

````python
data.info()
````

````python
# Check and drop NaN
data[data["PRCP"].isna()]
````

````python
data_clean = data.dropna()
````

````python
data_clean.info()
````

````python
X = data_clean[["MONTH","PRCP","TMAX","TMIN"]]
y = data_clean[["RAIN_SUB"]]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
````

````python
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
y_pred_log = logreg.predict(X_test)
logreg_score = logreg.score(X_test,y_test)
print(f"R-square : {logreg_score}")
train = pd.concat([X_test, y_test], axis='columns')
pred = pd.concat([train.reset_index(),pd.Series(y_pred_log, name='pred')], axis='columns')

pred
````
