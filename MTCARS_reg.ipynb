```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("mtcars.csv")
data
```

```python
data.info()
```

# Linear Regression Model



```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

X =data[["hp","wt"]]
y = data[["mpg"]]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
reg = LinearRegression().fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2 = r2_score(y_test,y_pred)
print(f"Intercept : {reg.intercept_}")
print(f"Coef : {reg.coef_}")
print(f"R-square :  {r2}")

```

```python
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(y_test, y_pred,squared=False)
print(f"RMSE : {rmse}")
```

```python
from sklearn.linear_model import Ridge,Lasso
score = []
for alpha in [0.1,1.0,10.0,100.0,1000.0]:
  ridge = Ridge(alpha = alpha)
  ridge.fit(X_train,y_train)
  y_pred_ridge = ridge.predict(X_test)
  score.append(ridge.score(X_test,y_test))
print(f"alpha : {score}")

```

```python
score2 = []
for alpha in [0.1,1.0,10.0,100.0,1000.0]:
  lasso = Lasso(alpha = alpha)
  lasso.fit(X_train,y_train)
  y_pred_lasso =lasso.predict(X_test)
  score2.append(lasso.score(X_test,y_test))

print(f"alpha : {score2}")
```

# Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X = data[["gear","mpg"]]
y = data[["am"]]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
logreg = LogisticRegression().fit(X_train,y_train)
y_pred_log = logreg.predict(X_test)
logreg_score = logreg.score(X_test,y_test)
print(f"R-square : {logreg_score}")
train = pd.concat([X_test, y_test], axis='columns')
pred = pd.concat([train.reset_index(),pd.Series(y_pred_log, name='pred')], axis='columns')

pred
```
